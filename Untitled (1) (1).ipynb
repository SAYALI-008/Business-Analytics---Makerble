```python
df = data
results = []
for var in categorical_vars:
    for mapped_var in mapped_vars:
        contingency_table = pd.crosstab(df[var], df[mapped_var])
        chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
        # Determine if the null hypothesis can be rejected
        hypothesis_result = "Reject Null Hypothesis" if p < 0.05 else "Fail to Reject Null Hypothesis"
    
    # Store results in a dictionary
        results.append({'Variable': var,
            'Chi-Square Statistic': chi2,
            'P-value': p,
            'Hypothesis Result': hypothesis_result,
            'Null Hypothesis': f"{mapped_var} is independent of {var}",
            'Alternative Hypothesis': f"{mapped_var} is dependent on {var}"
            })
results_df = pd.DataFrame(results)

# Filter for dependencies only (where the null hypothesis is rejected)
dependencies_df = results_df[results_df['Hypothesis Result'] == "Reject Null Hypothesis"]

# Print the clubbed results for dependencies
print("\nChi-Square Test Results (Dependencies Only):")
print(dependencies_df[['Variable', 'Chi-Square Statistic', 'P-value', 'Alternative Hypothesis']])
```

```python
import pandas as pd
import scipy.stats as stats
from itertools import combinations

# Get unique job titles
job_titles = data['Job Title'].unique()

# Prepare to store results
significant_results = []

# Loop through all combinations of job titles
for job1, job2 in combinations(job_titles, 2):
    # Filter data for the two job titles
    satisfaction_job1 = data[data['Job Title'] == job1]['Job Satisfaction']
    satisfaction_job2 = data[data['Job Title'] == job2]['Job Satisfaction']
    
    # Perform an independent t-test
    t_stat, p_value = stats.ttest_ind(satisfaction_job1, satisfaction_job2, nan_policy='omit')
    
    # Store results only if p-value is significant (less than alpha)
    alpha = 0.05
    if p_value < alpha:
        significant_results.append({
            'Job Title 1': job1,
            'Job Title 2': job2,
            'T-Statistic': t_stat,
            'P-Value': p_value
        })

# Convert results to a DataFrame for better presentation
significant_results_df = pd.DataFrame(significant_results)

# Display the final significant results
if significant_results_df.empty:
    print("No significant differences in Job Satisfaction between any job title pairs.")
else:
    print("Significant differences in Job Satisfaction between the following job title pairs:")
    print(significant_results_df)
```

```python
# Define the job titles
job_title1 = 'Product Manager'
job_title2 = 'UX Designer'

# Extract satisfaction scores for each job title
satisfaction_job1 = data[data['Job Title'] == job_title1]['Job Satisfaction']
satisfaction_job2 = data[data['Job Title'] == job_title2]['Job Satisfaction']

# Calculate Cohen's d
def cohen_d(group1, group2):
    """Calculate Cohen's d for two groups."""
    diff_mean = abs(group1.mean() - group2.mean())
    pooled_sd = (len(group1) * group1.std() ** 2 + len(group2) * group2.std() ** 2) / (len(group1) + len(group2))
    return diff_mean / pooled_sd ** 0.5

# Calculate Cohen's d for the specific job title pair
d = cohen_d(satisfaction_job1, satisfaction_job2)

print(f"Cohen's d: {d:.5f}")
```

```python

# List of numerical variables to analyze
numerical_variables = ['Job Satisfaction', 'Work-Life Balance', 'Management Support', 'Team Collaboration']

# List of categorical variables to analyze
categorical_variables = ['Age Bracket', 'Gender', 'Ethnicity', 'Job Title', 'Department']

# Prepare to store results
all_significant_results = {}

# Loop through each categorical variable
for categorical_variable in categorical_variables:
    significant_results = []

    # Get unique categories for the current categorical variable
    categories = data[categorical_variable].unique()

    # Loop through all combinations of categories
    for category1, category2 in combinations(categories, 2):
        # Filter data for the two categories
        for numerical_variable in numerical_variables:
            group1 = data[data[categorical_variable] == category1][numerical_variable].dropna()
            group2 = data[data[categorical_variable] == category2][numerical_variable].dropna()

            # Skip if either group is empty
            if len(group1) == 0 or len(group2) == 0:
                continue

            # Check for zero variance in either group to avoid division by zero
            if group1.var() == 0 or group2.var() == 0:
                continue  # Skip this comparison as T-statistic would be inf or -inf

            # Perform an independent t-test
            t_stat, p_value = stats.ttest_ind(group1, group2, nan_policy='omit')

            # Store results only if p-value is significant (less than alpha)
            alpha = 0.05
            if p_value < alpha:
                significant_results.append({
                    'Category 1': category1,
                    'Category 2': category2,
                    'Variable': numerical_variable,
                    'T-Statistic': t_stat,
                    'P-Value': p_value
                })

    # Convert results to a DataFrame for better presentation
    significant_results_df = pd.DataFrame(significant_results)

    # Store results for this categorical variable
    all_significant_results[categorical_variable] = significant_results_df

# Display the final significant results for each categorical variable
for categorical_variable, results_df in all_significant_results.items():
    print(f"\nSignificant differences in {categorical_variable} for the following comparisons:")
    if not results_df.empty:  # This checks if there are any significant results
        print(results_df)
    else:
        print(f"No significant differences in {categorical_variable} for any comparisons.")
```
